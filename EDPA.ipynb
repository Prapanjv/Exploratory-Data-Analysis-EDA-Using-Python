```python
# Function for log transformation of the column
def log_transform(edpa,col):
    for colname in col:
        if (edpa[colname] == 1.0).all():
            edpa[colname + '_log'] = np.log(edpa[colname]+1)
        else:
            edpa[colname + '_log'] = np.log(edpa[colname])
    edpa.info()
```
